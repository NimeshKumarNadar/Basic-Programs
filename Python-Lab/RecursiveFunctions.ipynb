# Recursive Functions

1. Factorial of a positive integer
2. Fibonacci number of given index
3. Tower's of Hanoi of 'n' disks
4. GCD of two positive integers
5. Sum of digits of a number
6. Reverse of a number
7. Decimal to binary
8. Decimal to octa
9. Decimal to Hexa
10. Binary to decimal
11. Binary to hexa
12. Sum of first 'n' natural numbers

```python
# 1. Factorial of a positive integer using recursion
def factorial(n):
    if(n==0):
        return 1
    else:
        return n*factorial(n-1)

# 2. Fibonacci number of given index using recursion
def fibonacci(n1,n2,n):
    if n!=0:
        print(n1+n2,end=" ")
        fibonacci(n2,n1+n2,n-1)
        
# 3. Tower's of Hanoi of 'n' disks
def ToH(n , src, dst, aux):
    if n==1:
        print ("Move disk 1 from source",src,"to destination",dst)
        return
    ToH(n-1, src, aux, dst)
    print ("Move disk",n,"from source",src,"to destination",dst)
    ToH(n-1, aux, dst, src)

# 4. GCD of two positive integers
def GCD(n1,n2,i):
    if(n1%i==0 and n2%i==0):
        return i
    else:
        return GCD(n1,n2,i-1)

# 5. Sum of digits of a number
def sumDigits(n):
    sum1=0
    if(n!=0):
        val = sumDigits(n//10)
        temp=n%10
        sum1=sum1+temp+val
    return sum1

# 6. Reverse of a number
def revNum(n):
    if(n!=0):
        print(n%10,end="")
        revNum(n//10)

# 7. Decimal to binary
def DecimalToBinary(n):
    if n!=0:
        DecimalToBinary(n//2)
        print(n%2,end="")

# 8. Decimal to octa
def DecimalToOctal(n):
    if n!=0:
        DecimalToOctal(n//8)
        print(n%8,end="")

# 9. Decimal to Hexa
def DecimalToHexa(n):
    if n!=0:
        DecimalToHexa(n//16)
        num=n%16
        if(num==10):
            print('A',end="")
        elif(num==11):
            print('B',end="")
        elif(num==12):
            print('C',end="")
        elif(num==13):
            print('D',end="")
        elif(num==14):
            print('E',end="")
        elif(num==15):
            print('F',end="")
        else:
            print(num,end="")

# 10. Binary to decimal
def BinaryToDecimal(n,i):
    if(n!=0):
        val = BinaryToDecimal(n//10,i+1)
        temp=n%10
        sum=val+temp*pow(2,i)
        return sum
    else:
        return 0

# 11. Binary to hexa
def BinaryToHexa(n):
    num=BinaryToDecimal(n,0)
    DecimalToHexa(num)

# 12. Sum of first 'n' natural numbers
def sumNat(n):
    if n>0:
        return n+sumNat(n-1)
    else:
        return 0

def main():
    print("List of recursive programs")
    print("1. Factorial of a positive integer")
    print("2. Fibonacci number of given index")
    print("3. Tower's of Hanoi of 'n' disks")
    print("4. GCD of two positive integers")
    print("5. Sum of digits of a number")
    print("6. Reverse of a number")
    print("7. Decimal to binary")
    print("8. Decimal to octa")
    print("9. Decimal to Hexa")
    print("10. Binary to decimal")
    print("11. Binary to hexa")
    print("12. Sum of first 'n' natural numbers")
    choice=1
    while(choice):
        choice = int(input("\n\nEnter a choice : "))
        if(choice==1):
            n=int(input("Enter number to find factorial : "))
            print("Factorial = ",factorial(n))
            
        elif(choice==2):
            n=int(input("Enter index of Fibonacci Series : "))
            print("0 1",end=" ")
            fibonacci(0,1,n-2)
            print("\n")
            
        elif(choice==3):
            n=int(input("No. of disks in Tower of Hanoi : "))
            ToH(n,'A','B','C')
            print("\nTotal steps = ",pow(2,n)-1)
            
        elif(choice==4):
            print("Input two numbers to find GCD : ")
            n1=int(input("First number : "))
            n2=int(input("Second number : "))
            if n1>n2:
                print("GCD = ",GCD(n1,n2,n2))
            else:
                print("GCD = ",GCD(n2,n1,n1))
            
        elif(choice==5):
            n=int(input("Enter number to find sum of digits : "))
            print("Sum of digits = ",sumDigits(n))
        
        elif(choice==6):
            n=int(input("Enter number to reverse : "))
            print("Reverse number = ",end="")
            revNum(n)
            
        elif(choice==7):
            n=int(input("Enter decimal number : "))
            print("Binary equivalent is : ",end="")
            DecimalToBinary(n)
            
        elif(choice==8):
            n=int(input("Enter decimal number : "))
            print("Octal equivalent is : ",end="")
            DecimalToOctal(n)
            
        elif(choice==9):
            n=int(input("Enter decimal number : "))
            print("Hexadecimal equivalent is : ",end="")
            DecimalToHexa(n)
            
        elif(choice==10):
            n=int(input("Enter binary number : "))
            print("Decimal equivalent is : ",end="")
            print(BinaryToDecimal(n,0))
            
        elif(choice==11):
            n=int(input("Enter binary number : "))
            print("Hexadecimal equivalent is : ",end="")
            BinaryToHexa(n)
            
        elif(choice==12):
            n=int(input("Enter Nth natural number : "))
            print("Sum = ",sumNat(n))
            
    print("Exiting...")
main()
```
