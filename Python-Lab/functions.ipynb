```python
def function1(n):
    sum1=0
    while(n!=0):
        temp=n%10
        n=int(n/10);
        sum1+=temp
    print("Sum of digits = ",sum1)
def function2(n):
    sum1=0
    while(n!=0):
        temp=n%10
        n=int(n/10);
        sum1=sum1*10+temp
    print("Reverse number = ",sum1)
def function3(n):
    hold=n
    sum1=0
    while(n!=0):
        temp=n%10
        n=int(n/10);
        sum1=sum1*10+temp
    if(hold==sum1):
        print("It is a palindrome number")
    else:
        print("Not a palindrome number")
def function4(n):
    maxd=0
    while(n!=0):
        temp=n%10
        n=int(n/10);
        if(temp>maxd):
            maxd=temp
    print("Maximum digit is ",maxd)
def function5(n):
    mind=n
    while(n!=0):
        temp=n%10
        n=int(n/10);
        if(temp<mind):
            mind=temp
    print("Minimum digit is ",mind)
def function6(n):
    n1=n
    maxd=0
    while(n!=0):
        temp=n%10
        n=int(n/10);
        if(temp>maxd):
            maxd=temp
    maxd1=0
    while(n1!=0):
        temp=n1%10
        n1=int(n1/10);
        if(temp>maxd1 and temp<maxd):
            maxd1=temp
    print("Second largest digit is ",maxd1)
def function7(n):
    n1=n
    mind=n
    while(n!=0):
        temp=n%10
        n=int(n/10);
        if(temp<mind):
            mind=temp
    mind1=n1
    while(n1!=0):
        temp=n1%10
        n1=int(n1/10);
        if(temp<mind1 and temp>mind):
            mind1=temp
    print("Second minimum digit is ",mind1)
    
def function8(n):
    ascending = "".join(sorted(str(n)))
    print(int(ascending))
    
def function9(n):
    s = n
    b = len(str(n))
    sum1 = 0
    while n != 0:
        r = n % 10
        sum1 = sum1+(r**b)
        n = n//10
    if s == sum1:
        print("Armstrong number")
    else:
        print("Not an armstrong number")
    
def function10(n):
    str_n = str(n)
    str_swapped = str_n[-1] + str_n[1:-1] + str_n[0]
    print("Swapped number : ",int(str_swapped))
    
def main():
    num=eval(input("Enter a number : "))
    print("Choose a task :\n1. Sum of digits\n2. Reverse the number\n3. Check Palindrome or not\n4. Maximum digit of the number\n5. Minimum digit of the number\n6. Second maximum digit\n7. Second minimum digit\n8. Sort digits ascending order\n9. Check Armstrong number\n10. Swap first and last digit\n0. EXIT\n")
    choice=1
    while(choice):
        choice = int(input("Enter a choice : "))
        if(choice==1):
            function1(num)
        elif(choice==2):
            function2(num)
        elif(choice==3):
            function3(num)
        elif(choice==4):
            function4(num)
        elif(choice==5):
            function5(num)
        elif(choice==6):
            function6(num)
        elif(choice==7):
            function7(num)
        elif(choice==8):
            function8(num)
        elif(choice==9):
            function9(num)
        elif(choice==10):
            function10(num)
    print("Exiting...")
main()
```
