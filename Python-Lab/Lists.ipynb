```python
def traversal(l):
    print("\nList : ",l)
    
def reversal(l):
    l = l[::-1]
    print("Reversed list : ",l)
    
def SortList(l,n):
    for i in range(0,n-1) :
        for j in range(0,n-1) :
            if(l[j]>l[j+1]) :
                temp = l[j]
                l[j] = l[j+1]
                l[j+1] = temp
    print("Sorted list : ",l)
    
def MaxElement(l,n):
    max = -1
    for i in range(n):
        if(max < l[i]) :
            max = l[i]
    print("Maximum element = ",max)
    
def MinElement(l,n):
    min = l[0]
    for i in range(n):
        if(min > l[i]) :
            min = l[i]
    print("Minimum Element = ",min)
    
def RemoveDuplicates(l,n):
    l1 = []
    size = 1
    temp = 0
    l1.append(l[0])
    for i in range(n):
        for j in range(size):
            if(l[i]==l1[j]):
                temp = 1
            else:
                temp = 0
        if(temp == 0):
            l1.append(l[i])
            size = size + 1
    print("\nList after removal of duplicate element : ",l1)

def main():
    list1=[]
    size=int(input("Enter size of list : "))
    print("Enter elements : ")
    for i in range(size):
        num=int(input())
        list1.append(num)
    print("\nAvailable operations :\n1. Traverse list\n2. Reverse list\n3. Sort list\n4. Maximum element\n5. Minimum element\n6. Unique elements\n0. EXIT\n\n")
    choice=1
    while(choice):
        choice=int(input("\nEnter choice : "))
        if choice==1:
            traversal(list1)
        elif choice==2:
            reversal(list1)
        elif choice==3:
            SortList(list1,size)
        elif choice==4:
            MaxElement(list1,size)
        elif choice==5:
            MinElement(list1,size)
        elif choice==6:
            RemoveDuplicates(list1,size)
        elif choice!=0:
            print("Invalid choice")
    print("Exiting...")
main()
```
